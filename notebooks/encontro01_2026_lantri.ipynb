```python
age = ""
while True:
  age_input_str = input("Enter your age: ")
  if age_input_str == "":
    print("You did not enter your age. Please try again.")
    continue
  if not age_input_str.isdigit():
    print("Your age must contain only digits. Please try again.")
    continue
  age_int = int(age_input_str)
  if age_int < 0:
    print("Your age cannot be negative. Please try again.")
    continue
  age = age_int
  break
print(f"Your age is {age} years old.")
```
